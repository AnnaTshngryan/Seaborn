```python
import pandas as pd 
import seaborn as sns 
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.describe()
```

## Barplot

```python
sns.barplot(x= "sex", y= "survived", data = df)
```

## Countplot 

```python
sns.countplot(x="sex",data=df)
```

## Boxplot

```python
sns.boxplot(x="pclass", y="age", data=df,palette='rainbow')
```

```python
sns.boxplot(x="pclass", y="age", hue="survived",data=df, palette="coolwarm")
```

## Violinplot

```python
sns.violinplot(x="pclass", y="age", data=df,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="pclass", y="age", data=df,hue='sex',split=True,palette='Set1')
```

## Stripplot

```python
sns.stripplot(x="pclass", y="age", data=df,jitter=True)
```

```python
sns.stripplot(x="pclass", y="age", data=df,jitter=True,hue='sex',palette='Set1')
```
